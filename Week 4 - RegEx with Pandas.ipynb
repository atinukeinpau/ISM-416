```python
import re
import pandas as pd
import email
```

```python
emails = []
fh = open('./fradulent_emails/fradulent_emails.txt', encoding='utf-8', errors='ignore').read()
```

```python
contents = re.split('From r', fh)
contents.pop(0)
```

```python
for item in contents:
    email_dict = {}
    sender = re.search('From:.*', item)
    print(sender.group())
```

```python
for item in contents:
    email_dict = {}
    
    #getting the sender
    sender = re.search('From:.*', item)
    #print(sender.group())
    if sender is not None:
        s_email = re.search('\w\S*@.*\w', sender.group())
        s_name = re.search('\".*\"', sender.group())
    else:
        s_email = None
        s_name = None
    if s_email is not None:
        sender_email = s_email.group()
    else:
        sender_email = None
    #print(sender_email)
    email_dict['sender_email'] = sender_email
    #print(email_dict)
    if s_name is not None:
        sender_name = s_name.group()
    else:
        sender_name = None
    email_dict['sender_name'] = sender_name
    #print(email_dict)
    
    #getting the recipient
    recipient = re.search('To:.*', item)
    if recipient is not None:
        r_email = re.search('\w\S*@.*\w', recipient.group())
        r_name = re.search('\".*\"', recipient.group())
    else:
        r_email = None
        r_name = None
    if r_email is not None:
        recipient_email = r_email.group()
    else:
        recipient_email = None
    email_dict['recipient_email'] = recipient_email
    if r_name is not None:
        recipient_name = r_name.group()
    else:
        recipient_name = None
    email_dict['recipient_name'] = recipient_name
    #print(email_dict)
    
    #Getting the date
    date_field = re.search("Date:.*", item)
    if date_field is not None:
        date = re.search("\d+\s\w+\s\d+", date_field.group())
    else:
        date = None
    if date is not None:
        date_sent = date.group()
    else:
        date_sent = None
    email_dict['date_sent'] = date_sent
    #print(email_dict)
    
    #getting the subject
    subject_field = re.search('Subject:.*', item)
    if subject_field is not None:
        subject = re.sub('Subject:', '', subject_field.group())
    else:
        subject = None
    email_dict['subject'] = subject
    #print(email_dict)
    
    #Getting email body
    full_email = email.message_from_string(item)
    #print(full_email.get_payload())
    body = full_email.get_payload()
    email_dict['email_body'] = body
    #print(email_dict)
    emails.append(email_dict)
```

```python
email_df = pd.DataFrame(emails)
```
