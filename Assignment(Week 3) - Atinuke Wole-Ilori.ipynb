### Question 1:
A regular expression (a.k.a RegEx) is a sequence of characters that specifies a search pattern. They are patterns that are used by string-searching algorithms for "find" or "find and replace" operations on strings, or for input validation.

```python
file = open('test_emails.txt').read()
```

```python
import re
```

### Question 2: 

```python
match = re.findall('Content-Type: .*', file)
```

```python
match
```

```python
for detail in match:
    print(re.findall("\s.+", detail))
```

### Question 3

```python
match2 = re.findall('Date:.*', file)
match2
```

```python
for date in match2:
    print(re.findall("\d+\s+\w+\s\d*", date))
```
