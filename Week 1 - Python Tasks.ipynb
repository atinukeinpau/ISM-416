```python
def maximum (a, b, c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    elif c > a and c > b:
        return c
    else: 
        return 'They are all equal.'
```

```python
maximum(1, 2, 3)
```

```python
maximum(105, 876, 180)
```

```python
maximum(1, 1, 1)
```

```python
def addup (x):
    total = 0
    for i in x:
        total = total + i
    return total
```

```python
addup([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
def mul (x):
    total = 1
    for i in x:
        total = total * i
    return total
```

```python
mul([1, 2, 3, 4, 5, 6, 7, 8, 9])
```
