4. One challenge of NLP is ambiguity which means a word or sentence could have two or more meanings. It can be categorised into three; lexical(a word that could be used as a verb, adjective or noun), syntatic(multiple interpretations of a sentence in context) and referential(a word refers to more than one element in the statement).
Another challenge of NLP is the detection of irony and sarcasm because some statements could be written to mean something but the author meant the opposite.
Contextual words and phrases and homonyms could also stand to be a challenge for NLP because some words have two different meanings but have the same spelling and pronunciation and there could be a problem with differentiation.
